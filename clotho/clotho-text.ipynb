# Clotho translated (Thai)

```python
import json
```

```python
def read_jsonl(file_path):
    data = []
    with open(file_path, 'r') as file:
        for line in file:
            data.append(json.loads(line.strip()))
    return data
```

```python
train.iloc[0]['file_name']
```

```python
paths = [
    "/workspace/translate-data/outputs/clotho/train/0_2000.jsonl",
    "/workspace/translate-data/outputs/clotho/train/2000_3839.jsonl",
]
train_th = []
for path in paths:
    train_th += read_jsonl(path)
print(len(train_th))


dataset = []
for i in range(len(train)):
    x = train.iloc[i]
    assert train_th[i]['file_name'].split('/')[-1] == x['file_name']

    text1 = train_th[i]['caption_1_th'].split("<|im_start|>")[0] 
    text2 = train_th[i]['caption_2_th'].split("<|im_start|>")[0] 
    text3 = train_th[i]['caption_3_th'].split("<|im_start|>")[0] 
    text4 = train_th[i]['caption_4_th'].split("<|im_start|>")[0] 
    text5 = train_th[i]['caption_5_th'].split("<|im_start|>")[0] 
    
    dataset.append({
        'path': f"/dataset/audio-data/clotho/wav_16k/development/{x['file_name']}",
        'text_1': text1,
        'text_2': text2,
        'text_3': text3,
        'text_4': text4,
        'text_5': text5,
        
    })
with open("./train_th.json", "w") as f:
    json.dump(dataset, f, indent=4, ensure_ascii=False)
```

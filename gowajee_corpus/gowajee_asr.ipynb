# Process text

```python
from glob import glob
import json
```

```python
path = "v0.9.2/train/text"
with open(path) as f:
    lines = f.readlines()
lines = [line.strip() for line in lines]
data = []
for line in lines:
    file_name = line[:16]

    path = glob(f"/dataset/audio-data/gowajee_corpus/v0.9.2/audios/*/{file_name}.wav")
    assert len(path) == 1
    path = path[0]
    
    text = line[17:]
    text = text.split(" ")
    text = "".join(text)
    data.append({
        'path': path,
        'text': text,
    })
print(len(data))
with open("./train.json", "w") as f:
    json.dump(data, f, indent=4, ensure_ascii=False)
```

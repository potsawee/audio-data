# Prepare training data (En -> Th)

```python
from glob import glob
from tqdm import tqdm
import os
import json
import random
```

```python
def read_jsonl(file_path):
    data = []
    with open(file_path, 'r') as file:
        for line in file:
            json_object = json.loads(line)
            data.append(json_object)
    return data
```

```python
# dir_path = "./text_with_thai/en_de.train/"
def read_dir(dir_path):
    paths = sorted(glob(f"{dir_path}/*.jsonl"))
    data = []
    for path in paths:
        data += read_jsonl(path)
    print("len", len(data))
    return data
```

```python
data = read_dir("./text_with_thai/en_de.train/")
```

```python
len(data)
```

```python
data[0]
```

```python
examples = []
for x in tqdm(data):
    path = x['path'].replace("mp3_original", "mp3_16kHz")
    assert os.path.isfile(path)
    examples.append({
        'path': path,
        'task': "translation_en2th",
        'text': x['thai'],
    })
```

```python
with open("./train.en2th.json", "w") as f:
    json.dump(examples, f, indent=4, ensure_ascii=False)
```
